### Task1: Use Gaussian Process to find a good function that passes exatcly through the points (x, y)

```python
from sklearn.gaussian_process import GaussianProcessRegressor
import numpy as np
import matplotlib.pylab as plt
```

```python
gpr = GaussianProcessRegressor()
gpr.fit(x, y)

x_test = np.linspace(10,50, 1000).reshape(-1, 1)
y_pred, std_prediction= gpr.predict(x_test, return_std=True)

plt.figure(figsize=(15,15))
plt.plot(x_test, y_pred)
plt.scatter(x, y, c='red')
plt.fill_between(
    x_test.ravel(),
    y_pred.ravel() - 1.96 * std_prediction,
    y_pred.ravel() + 1.96 * std_prediction,
    alpha=0.5,
    label=r"95% confidence interval",
)
plt.show()
```

### Task2: Plot the posterior mean function and a sample function from the posterior distribution

```python
from sklearn.gaussian_process.kernels import ExpSineSquared, RBF
kernel = RBF() + RBF() * ExpSineSquared(1,1,periodicity_bounds=(1,3))

gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=200, normalize_y=True)
gpr.fit(x, y)

x_test = np.linspace(10, 50, 1000).reshape(-1, 1)
y_pred, std_prediction= gpr.predict(x_test, return_std=True)

plt.figure(figsize=(15,15))
plt.plot(x_test, y_pred)
plt.scatter(x, y, c='red')
plt.fill_between(
    x_test.ravel(),
    y_pred.ravel() - 1.96 * std_prediction,
    y_pred.ravel() + 1.96 * std_prediction,
    alpha=0.5,
    label=r"95% confidence interval",
)
```

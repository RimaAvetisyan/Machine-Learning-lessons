```python
def preprocess(x):
    for i in range(1, 7):
        dx = f'dx{i}'
        x[dx] = x['Treated_with_drugs'].apply(lambda x: int(type(x) == str and dx in x.lower()))
    x['age_mass'] = x['Patient_Age'] * x['Patient_Body_Mass_Index']
    x['age_mass2'] = x['age_mass'] * x['Patient_Body_Mass_Index']
    return x
```

```python
x = preprocess(x)
```

```python
x = x[x_cols]
```

```python
from sklearn.model_selection import train_test_split
```

```python
for c in x.columns:
    col_type = x[c].dtype
    if col_type == 'object' or col_type.name == 'category':
        x[c] = x[c].astype('category')
X_train, X_valid, y_train, y_valid = train_test_split(x, y, test_size=0.05)
```

```python
X_train.head()
```

```python
clf = lgb.LGBMClassifier(n_estimators=1000, learning_rate=0.1)
```

```python
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, accuracy_score

predictions = clf.predict(X_valid)
print("valid accuracy_score: " + str(accuracy_score(predictions, y_valid)))
```

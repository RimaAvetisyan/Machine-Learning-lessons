## T-SNE on MNIST dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Task1: Read the data from "mnist_train.csv" file using pandas and separate the 'label' column to use later for coloring

```python
import pandas as pd
mnist_df = pd.read_csv('mnist_train.csv')
```

```python
mnist_df.head()
```

```python
labels = mnist_df.pop('label')
```

```python
labels
```

```python
mnist_df.head()
```

### Task 2: Perform PCA to reduce the dimension to 2 for plotting

```python
from sklearn.decomposition import PCA
pca = PCA(2)
X_2d = pca.fit_transform(mnist_df)
```

### Task 3: Plot the two dimensional points resulting from the PCA

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))
plt.scatter(X_2d[:,0], X_2d[:,1], c=labels, s=5)
plt.show()
```

### Task 4: Perform T-SNE to reduce the dimension to 2 for plotting
Note: it might take minutes with default parameters and the full data. If it takes too long you can try using part of the data

```python
from sklearn.manifold import TSNE
tsne = TSNE()
Xtsne = tsne.fit_transform(mnist_df)
```

### Task 5: Plot the two dimensional points resulting from the T-SNE

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,15))
plt.scatter(Xtsne[:,0], Xtsne[:,1], c=labels, s=2, cmap='tab10')
plt.colorbar()
plt.show()
```

### Optional: Try different parameters, try to change the perplexity

```python
from sklearn.manifold import TSNE
tsne = TSNE(perplexity=20)
Xtsne = tsne.fit_transform(mnist_df.iloc[:10000])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,15))
plt.scatter(Xtsne[:,0], Xtsne[:,1], c=labels[:10000], s=2, cmap='tab10')
plt.colorbar()
plt.show()
```
